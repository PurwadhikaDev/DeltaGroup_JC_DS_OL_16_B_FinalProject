Analisis Kesalahan Prediksi:

Tipe 1 : *False Positive* (FP)
Jika yang terjadi adalah FP, dimana model memprediksi pemesanan dibatalkan padahal aktualnya pemesanan tidak dibatalkan, maka risiko yang diterima oleh pihak hotel adalah:
* marketing yang tidak tepat sasaran, sehingga mengeluarkan biaya marketing yang sia-sia

Tipe 2 : *False Negative* (FN)
Jika yang terjadi adalah FN, dimana model memprediksi pemesanan tidak dibatalkan padahal aktualnya pemesanan dibatalkan, maka risiko yang diterima oleh pihak hotel adalah:
* kehilangan pendapatan
* sia-sianya biaya, waktu, dan sumber daya untuk merealisasikan ekspektasi calon penghuni

Analisis *Metric* yang Akan Digunakan:

Dari analisis risiko atas terjadinya kesalahan prediksi, diperoleh kerugian yang lebih besar akan diterima oleh pihak hotel jika terjadi kesalahan prediksi tipe 2, yaitu *False Negative* (FN), dimana model memprediksi pemesanan tidak dibatalkan padahal aktualnya pemesanan dibatalkan. Oleh karena itu, untuk mereduksi terjadinya kesalahan prediksi tersebut, digunakan *metric F2 Score*.

*F2 Score* adalah salah satu metric yang menitikberatkan pada nilai recall daripada precision. *Metric* ini digunakan jika mereduksi FN menjadi prioritas tanpa meninggalkan FP. Rumus *F2 Score* :

## **Data Correlation & Features Selection**

Pada langkah ini, akan dilakukan pengecekan hubungan setiap kolom dengan kolom target. Kolom-kolom yang berhubungan erat dengan kolom target akan dijadikan sebagai feature, sehingga kolom yang hubungannya tidak kuat dengan kolom target akan didrop.

Tahap ini akan dipisah menjadi dua, yaitu:

- Cek hubungan kolom kategorik dengan kolom target
- Cek hubungan kolom numerik dengan kolom target

Tahap tersebut dipisah menjadi dua karena dilakukan dengan metode yang berbeda.

### **Cek Korelasi Keputusan Pembatalan dengan Kolom Numerik**

Pengecekan korelasi data numerik dapat dilakukan jika semua kolom yang akan dicek berbentuk numerik.

Suatu kolom dikatakan berkorelasi dengan kolom lain, dalam hal ini adalah kolom target, jika nilai korelasinya lebih dari atau sama dengan 10%, sehingga jika nilai korelasinya di bawah 10%, kolom tersebut akan didrop atau tidak dijadikan sebagai feature.

### **Cek Korelasi Keputusan Pembatalan dengan Kolom Kategorik**

Untuk mengecek hubungan data kategorik dengan kolom target, kita gunakan metode chi square.

Jika p-value < 0,05 maka kolom tersebut mempunyai pengaruh terhadap kolom lainnya, dalam hal ini adalah kolom targetnya.

Karena pada dataset kolom 'is_canceled' berbentuk numerik, makan akan diubah terlebih dahulu isinya sebagai berikut.

- 1 : 'Yes', artinya pemesanan dibatalkan
- 0 : 'No', artinya pemesanan tidak dibatalkan

```python
df['is_canceled'].unique()
```

Karena kolom is_canceled masih berupa integer, maka akan diubah terlebih dahulu ke bentuk string agar bisa dihitung korelasinya dengan kolom-kolom kategorik menggunakan metode chi-square. Isi kolom tersebut akan diubah menjadi :
- 0 : 'No'
- 1 : 'Yes'

Artinya 'Yes' dijadikan sebagai target positif karena kita akan memprediksi apakah tamu akan membatalkan pesanannya, dimana ketika tamu membatalkan pesanannya, maka di kolom target akan ditulis 'Yes'.

- Kolom arrival_date_month, market_segment, assigned_room_type, agent, booking_change_category di-encoding menggunakan Binary Encoder karena banyak unique dari kolom-kolom tersebut lebih dari 5.

2. Untuk tipe data 'float' dan 'int'

Kolom is_local dan cancellations_before pada dasarnya adalah kolom kategorik, hanya saja dinyatakan dalam angka, sehingga kolom ini tidak perlu dilakukan encoding lagi dan juga tidak perlu dilakukan scalling.

Karena data tersebut sudah berupa numerik, maka tidak perlu dilakukan encoding, sehingga kita hanya perlu melakukan scalling. Kita akan menggunakan robust scaler untuk kolom numerik, yaitu lead_time, required_car_parking_spaces, total_of_special_requests menggunakan Robust Scaler karena Robust Scaler adalah metode scaling yang paling tahan terhadap outliers.

# **Preprocessing**

## **Definisikan X dan y**

- X adalah kolom feature
- y adalah kolom target

```python
# define x and y

X = df_final.drop('is_canceled', axis=1)
y = df_final['is_canceled']
```

## **Lakukan Split Data**

Data akan dibagi menjadi data train (data latih) dan data test (data uji) dengan perbandingan banyak data train dan data test adalah 80% : 20%.

```python
# Split Data (Train & Test)

xtrain, xtest, ytrain, ytest = train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state= 2024,
    stratify = y
)
```

## **Membuat Skema Preprocessing**

Pada tahap ini, akan dilakukan feature engineering, yaitu proses encoding dan scalling yang telah ditentukan sebelumnya.

```python
# Membuat skema preprocessing

one_hot_col = ['hotel','meal','distribution_channel','deposit_type','customer_type']
binary_en_col = ['market_segment', 'assigned_room_type', 'agent', 'arrival_date_month','booking_change_category']
rs_col = ['lead_time','adr','required_car_parking_spaces','total_of_special_requests']

preprocessing = ColumnTransformer(
    transformers = [
        ('OHE', OneHotEncoder(handle_unknown='ignore'), one_hot_col),
        ('BinaryEnc', ce.BinaryEncoder(),binary_en_col),
        ('RobustScal', RobustScaler(), rs_col)
    ], remainder = 'passthrough'
)

preprocessing
```

## **Membuat Model Machine Learning**

Karena kita akan memprediksi apakah seseorang tamu akan membatalkan pemesanan hotel atau tidak, maka kita gunakan model klasifikasi.

**Mendefinisikan Model yang Akan Digunakan**

```python
# Initialize Models

dt = DecisionTreeClassifier(random_state=2024)
knn = KNeighborsClassifier()
logreg = LogisticRegression(random_state=2024)
nb = GaussianNB()

hard_vote = VotingClassifier([
    ('knn',knn),
    ('LogReg', logreg),
    ('DT', dt),
    ('Naive Bayes', nb),
], voting='hard')

stacking = StackingClassifier([
    ('knn',knn),
    ('LogReg', logreg),
    ('DT', dt),
    ('Naive Bayes', nb),
], final_estimator=dt)

rf = RandomForestClassifier(random_state=2024)

adb = AdaBoostClassifier(random_state=2024)

bagging = BaggingClassifier(n_estimators=300, random_state=2024)

xgb = XGBClassifier()

gb = GradientBoostingClassifier(random_state=2024)

models = {
    'DecisionTree': dt,
    'KNN': knn,
    'Logistic Regression' : logreg,
    'Naive Beiyes' : nb,
    'Hard Vote' : hard_vote,
    'Stacking' : stacking,
    'Random Forest' : rf,
    'Adaboost' : adb,
    'Bagging' : bagging,
    'XG Boost' : xgb,
    'Gradient Boost' : gb
}
```

**Membuat Pipeline Model Tanpa Balancing Data Target**

Proses ini dilakukan untuk melihat f2 score jika data tidak diseimbangkan lebih dulu, mengingat data target yang jumlahnya tidak seimbang.

```python
# Create Pipeline model tanpa Balancing

f2_scorer = make_scorer(fbeta_score, beta=2)
hasil = []
namaModel = []

for model in models.items() :
    model_pipe = Pipeline(steps = [
        ('Pre Process', preprocessing),
        ('Model', model[1])
    ])

    cv_score = cross_val_score(model_pipe, xtrain, ytrain, cv=5, scoring = f2_scorer)

    hasil.append({
        'Model' : model[0],
        'F2 Score' : cv_score,
        'AVG F2 Score' : cv_score.mean(),
        'Std F2 Score' : cv_score.std()
    })

pd.DataFrame(hasil).sort_values('AVG F2 Score', ascending = False)
```

### **Definisikan Metode Balancing pada Data Target**

Karena data target jumlahnya tidak seimbang, maka diperlukan balancing pada data target.

```python
# Definisikan balancing methode
balancing_methods = {
    'SMOTE': SMOTE(random_state=2024),
    'ADASYN': ADASYN(random_state=2024),
    'RandomOverSampler': RandomOverSampler(random_state=2024),
    'RandomUnderSampler': RandomUnderSampler(random_state=2024),
    'NearMiss': NearMiss(version=3),
    'BorderlineSMOTE' : BorderlineSMOTE(random_state = 2024)
}
```

### **Membuat Pipeline Model dengan Balancing Data Target**

Pada tahap ini, kita akan bandingkan recall score untuk setiap model. Lima model dengan F2 score terbaik, akan diambil untuk selanjutnya dilakukan hyperparameter tuning. Harapannya, dengan melakukan hyperparameter tuning, recall score akan meningkat dan memperbagus model untuk diterapkan ke data baru.

```python
# Create Pipeline model dengan Balancing

hasil = []
namaModel = []

for model in models.items() :
    for i in balancing_methods.items():
        model_pipe = Pipeline(steps = [
            ('Pre Process', preprocessing),
            ('Balancing', i[1]), 
            ('Model', model[1])
        ])

        cv_score = cross_val_score(model_pipe, xtrain, ytrain, cv=5, scoring = f2_scorer)

        hasil.append({
            'Model' : model[0],
            'Resample' : i[0],
            'F2 Score' : cv_score,
            'AVG F2 Score' : cv_score.mean(),
            'Std F2 Score' : cv_score.std()
        })
```

```python
pd.DataFrame(hasil).sort_values('AVG F2 Score', ascending = False)
```

Hyperparameter tuning akan dilakukan untuk lima model, yaitu:
1. XG Boost dengan metode balancing Random Under Sampler
2. Gradient Boost dengan metode balancing BorderlineSMOTE

Rekomendasi ini bertujuan untuk meningkatkan keakurasian dan ketepatan model dalam memprediksi pembatalan pemesanan, serta mengurangi kesalahan prediksi yang dapat berdampak pada pengambilan keputusan bisnis.

1. Mencoba algoritma boosting lain yang belum digunakan untuk melihat apakah dapat meningkatkan performa model.
2. Menambah jumlah data, terutama untuk kategori canceled, agar model memiliki lebih banyak informasi dan dapat membuat prediksi yang lebih akurat.
3. Menganalisis data yang masih salah diprediksi, terutama False Positive dan False Negative, untuk memahami pola kesalahan dan mencari cara memperbaikinya.
4. Menggunakan data yang lebih baru agar dapat lebih menggambarkan situasi dan kondisi saat ini.
